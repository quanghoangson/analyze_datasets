##                                         Netflix Movies and TV Shows Data Analysis

### 2.  Assess Clean data



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import matplotlib
```

```python
netflix_titles_df = pd.read_csv('netflix_titles.csv')
netflix_titles_df.head()
```

```python
netflix_titles_df.info()
```

```python
netflix_titles_df
```

```python
netflix_titles_df.nunique()
```


### 2.1 Handling Null Values


```python
netflix_titles_df.isnull().sum()
```

```python
netflix_titles_df['director'].fillna('No Director', inplace=True)
netflix_titles_df.dropna(subset=['date_added', 'rating','cast','country','duration'], inplace=True)
```

```python
netflix_titles_df.isnull().sum()
```

### 2.2 Splitting the Dataset


```python
netflix_movies_df = netflix_titles_df[netflix_titles_df['type'] == 'Movie'].copy()
netflix_movies_df.head()
```

```python
netflix_shows_df = netflix_titles_df[netflix_titles_df['type']=='TV Show'].copy()
netflix_shows_df.head()
```

### 2.3 Data Preparation

```python
netflix_movies_df['duration'] = netflix_movies_df['duration'].str.replace(' min', '')
netflix_movies_df['duration'] = netflix_movies_df['duration'].astype(int)
netflix_movies_df
```

```python
netflix_shows_df.rename(columns={'duration': 'seasons'}, inplace=True)
netflix_shows_df['seasons'] = netflix_shows_df['seasons'].str.replace(' Seasons', '').str.replace(' Season', '').astype(int)
netflix_shows_df
```

```python
netflix_titles_df_clean = netflix_titles_df.copy()
netflix_titles_df_clean.to_csv('netflix_titles_df_clean.csv')
netflix_titles_df_clean.isna().sum()
```

### 4. Visualizations & Communication 

```python
netflix_titles_df_clean.head()
```

```python
# Calculate the quantity of each type
type_counts = netflix_titles_df_clean['type'].value_counts()
type_counts
```

```python
plt.figure(figsize=(7,5))
g = sns.barplot(x=type_counts.index, y=type_counts.values, palette="pastel")
plt.title("Tổng Movies và TV Shows")
plt.ylabel("Số lượng")
plt.show()
```

```python

plt.figure(figsize=(12,6))
plt.title("% of Netflix Titles that are either Movies or TV Shows")
g = plt.pie(netflix_titles_df.type.value_counts(), explode=(0.025,0.025), labels=netflix_titles_df_clean.type.value_counts().index, colors=['skyblue','navajowhite'],autopct='%1.1f%%', startangle=180)
plt.legend()
plt.show()
```

#### Vậy là có khoảng hơn 4.000 bộ phim và gần 2.000 chương trình truyền hình với phim chiếm đa số. Tuy nhiên, khi xem xét theo tựa đề, có nhiều tựa phim hơn (68.5%) so với tựa chương trình truyền hình (31.5%).

```python

filtered_countries = netflix_titles_df_clean.set_index('title').country.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
filtered_countries = filtered_countries[filtered_countries != 'Country Unavailable']


plt.figure(figsize=(7,9))
g = sns.countplot(y=filtered_countries, order=filtered_countries.value_counts().index[:20])
plt.title('Top 20 Countries on Netflix')
plt.xlabel('Titles')
plt.ylabel('Country')
plt.show()

```

####  Chúng ta sẽ khám phá các quốc gia có nhiều nội dung nhất trên Netflix. Sau khi tách riêng các quốc gia và loại bỏ các tựa đề không có thông tin quốc gia, chúng ta có thể vẽ biểu đồ Top 20 để xem xét xem các quốc gia nào có sẵn nhiều bộ phim nhất trên Netflix. Không ngạc nhiên khi Hoa Kỳ đứng đầu với Netflix là một công ty Mỹ. Ấn Độ đến thứ hai , tiếp theo là Anh và Canada.

#### 4.3 Netflix Film Ratings


##### Qua biểu đồ ta có thể thấy được TV-MA đứng đầu trong bảng xếp hạng tiếp sau đó là kêng TV-14 

```python
order = ['G', 'TV-Y', 'TV-G', 'PG', 'TV-Y7', 'TV-Y7-FV', 'TV-PG', 'PG-13', 'TV-14', 'R', 'NC-17', 'TV-MA']
plt.figure(figsize=(15,7))
g = sns.countplot(x=netflix_titles_df.rating, hue=netflix_titles_df_clean.type, order=order, palette="pastel");
 
plt.title("Ratings for Movies & TV Shows")
plt.xlabel("Rating")
plt.ylabel("Total Count")
plt.show()
```

```python
rating_counts = netflix_titles_df['rating'].value_counts().reindex(order).fillna(0)

plt.figure(figsize=(10, 7))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel", len(order)))
g.bar_label(g.containers[0])  
plt.title("Ratings for Movies & TV Shows")
plt.show()
```

#### 4.4 Popular Genres


```python
filtered_genres = netflix_titles_df_clean.set_index('title').listed_in.str.split(', ', expand=True).stack().reset_index(level=1, drop=True);

plt.figure(figsize=(7,9))
g = sns.countplot(y = filtered_genres, order=filtered_genres.value_counts().index[:20])
plt.title('Top 20 Genres on Netflix')
plt.xlabel('Titles')
plt.ylabel('Genres')
plt.show()
```

#### Trong mảng thể loại, phim quốc tế đứng đầu một cách bất ngờ, tiếp theo là phim tình cảm và phim hài. Mặc dù Hoa Kỳ có nhiều nội dung nhất, nhưng có vẻ Netflix đã quyết định phát hành một lượng lớn phim quốc tế. Lý do cho điều này có thể là do hầu hết các thuê bao của Netflix thực sự không ở trong Hoa Kỳ, mà thay vào đó, đa số khán giả thực sự là các thuê bao quốc tế.

#### 4.5 Distribution of TV Shows by Seasons

```python
tv_shows = netflix_titles_df_clean[netflix_titles_df["type"] == "TV Show"].copy()
tv_shows.loc[:, "Seasons"] = tv_shows["duration"].apply(lambda x: int(x.split()[0]) if "Season" in x else None)
plt.figure(figsize=(10, 6))
sns.histplot(tv_shows["Seasons"].dropna(), bins=30, color="blue", kde=True)
plt.title("Distribution of TV Shows by Seasons")
plt.xlabel("Seasons")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

#### Biểu đồ sử dụng histogram để hiển thị phân phối, trong đó mỗi cột đại diện cho một khoảng của số lượng mùa và chiều cao của cột thể hiện số lượng chương trình truyền hình có số mùa trong khoảng đó.

#### 4.6 Distribution of Movies by Duration

```python
movies = netflix_titles_df_clean[netflix_titles_df_clean["type"] == "Movie"].copy()
movies.loc[:, "Duration"] = movies["duration"].apply(lambda x: int(x.split()[0]) if isinstance(x, str) and "min" in x else None)
movies = movies.dropna(subset=["Duration"])
plt.figure(figsize=(10, 6))
sns.histplot(movies["Duration"], bins=30, color="red", kde=True)
plt.title("Distribution of Movies by Duration")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

```

####  Biểu đồ sử dụng histogram để hiển thị phân phối, trong đó mỗi cột đại diện cho một khoảng thời lượng và chiều cao của cột thể hiện số lượng phim trong khoảng thời lượng đó.
